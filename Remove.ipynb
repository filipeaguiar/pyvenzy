# Alterar todos os para inverter CardNumber e ClearCode

```python
import requests
import json
import settings
import warnings
import pandas as pd
from paciente import Paciente

warnings.filterwarnings("ignore")
settings.init()
```

## Obter todos os Cart√µes

```python
df = pd.read_csv('ativos.csv')
pacientes = df[df['CHType'] == 2]

pacientes_list = pacientes['CHID'].tolist()

total = 0

for paciente in pacientes_list:
    user = requests.get(
        url=f'{settings.baseUrl}/cardholders/{paciente}',
        headers={"Content-Type": "application/json"},
        verify=False,
    )
    user_json = user.json()
    user_json['CHState'] = 1
    response = requests.put(
        url=f'{settings.baseUrl}/cardholders/',
        headers={"Content-Type": "application/json"},
        data=json.dumps(user_json),
        verify=False,
        params=(("CallAction", True),)
    )
    total += 1
    print(response.status_code)
    
print(str(total))
```
