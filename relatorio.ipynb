```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('relatorio.csv', sep=';')

cardgroup = df['Card'].value_counts()
activegroup = df['CardStat'].value_counts()
```

```python
print(cardgroup)
print(activegroup)
```

```python
plt.figure(figsize=(10, 5))
plt.pie(cardgroup, labels=cardgroup.index.to_list(), autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Quantidade de cart√µes por tipo')
plt.axis('equal')
plt.show()
plt.figure(figsize=(10, 5))
plt.pie(activegroup, labels=activegroup.index.to_list(), autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Quantidade de cart√µes por status')
plt.axis('equal')
plt.show()
```
